```python
import tensorflow as tf
import numpy as np
from scipy import misc
from matplotlib import pyplot as plt
%matplotlib inline

sess = tf.Session()

image = misc.imread('Vd-Orig.png')
plt.imshow(image, cmap='gray')
```

```python
sess = tf.Session()

input = tf.expand_dims(tf.cast(image, tf.float32), 0)
filter_ = np.random.random(9).reshape([1, 1, 3, 3])  # filter_height, filter_width, in_channels, out_channels
output = tf.nn.conv2d(input, filter_, strides=[1, 1, 1, 1], padding='SAME')

print(input)
print(filter_)
print(output)

im2 = tf.cast(tf.squeeze(output), tf.uint8)
plt.imshow(sess.run(im2))
```

```python
filter_ = np.random.random(3*3*3*3).reshape([3, 3, 3, 3])
output = tf.nn.conv2d(input, filter_, strides=[1, 1, 1, 1], padding='SAME')
print(output)

im3 = tf.cast(tf.squeeze(output), tf.uint8)
plt.imshow(sess.run(im3))
```
